```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
station_cols = ['MonitoringLocationIdentifier', 'LatitudeMeasure', 'LongitudeMeasure']
station_nums = list(range(1, 5))
data = pd.DataFrame(columns = station_cols)

for i in station_nums:
    data_new = pd.read_csv(f'station{i}.csv')
    data_new = data_new[station_cols]
    data = data.append(data_new)
data = data.set_index('MonitoringLocationIdentifier')
```

```python
chem_cols = ['MonitoringLocationIdentifier', 'CharacteristicName',
             'ResultMeasure/MeasureUnitCode', 'ResultMeasureValue',
             'ActivityStartDate', 'ActivityStartTime/Time',
             'ActivityStartTime/TimeZoneCode']
data2 = pd.read_csv('chemical_data.csv', usecols = chem_cols)
data2 = data2.set_index('MonitoringLocationIdentifier')
```

```python
results = data.merge(data2, left_index = True, right_index = True, how='outer')
```
